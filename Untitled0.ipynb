```python
import sys, math
import numpy as np
from scipy import stats

def field_equation(pi, cost_ratio, n, fidelity):
    raw = (pi / (1 - pi)) * cost_ratio * (1 / math.sqrt(n)) * fidelity
    return min(1.0, raw)

def robust_alpha(pi, cost_ratio, n, fidelity,
                 sigma_pi=0.30, sigma_C=0.35, sigma_n=0.04, sigma_Phi=0.10,
                 rho_pi_Phi=0.0, gamma=0.3, q=0.05):
    log_mean = (math.log(pi/(1-pi)) + math.log(cost_ratio)
                - 0.5*math.log(n) + math.log(fidelity))
    A_sq = (sigma_pi**2 + sigma_C**2 + sigma_n**2 + sigma_Phi**2
            + 2*rho_pi_Phi*sigma_pi*sigma_Phi)
    A = math.sqrt(A_sq)
    z = float(stats.norm.ppf(1 - q))
    dro = math.sqrt(q / (1 - q))
    sf = (1 - gamma)*z + gamma*dro
    return math.exp(log_mean - sf * A)

# RECOVERY IMV
at = field_equation(0.52, 6.67, 1028, 0.87)
ar = robust_alpha(0.52, 6.67, 1028, 0.87, rho_pi_Phi=0.40)
print(f"IMV:  alpha_target={at:.3f}  alpha_robust={ar:.3f}")

# RECOVERY NoRS
at2 = field_equation(0.12, 0.35, 1542, 0.84)
ar2 = robust_alpha(0.12, 0.35, 1542, 0.84)
print(f"NoRS: alpha_target={at2:.5f}  alpha_robust={ar2:.5f}")
```

```python
subgroups = [
    ("IMV",         0.52, 6.67, 1028, 0.87, 0.40, "< 0.001", "TREAT"),
    ("O2 only",     0.38, 2.00, 3855, 0.86, 0.00, "0.004",   "TREAT"),
    ("No support",  0.12, 0.35, 1542, 0.84, 0.00, "~0.15",   "DO NOT TREAT"),
]

print(f"{'Subgroup':<14} {'π':>5} {'C2/C1':>6} {'n':>6} {'Φ':>5} {'α_target':>10} {'α_robust':>10} {'Obs.p':>8}  Verdict")
print("-" * 95)

for label, pi, cr, n, phi, rho, obs_p, verdict in subgroups:
    at = field_equation(pi, cr, n, phi)
    ar = robust_alpha(pi, cr, n, phi, rho_pi_Phi=rho)
    print(f"{label:<14} {pi:>5.2f} {cr:>6.2f} {n:>6} {phi:>5.2f} {at:>10.5f} {ar:>10.5f} {obs_p:>8}  {verdict}")

print()
print(f"Fixed α=0.05 is {0.05/field_equation(0.12,0.35,1542,0.84):.0f}x MORE PERMISSIVE than FWAS for NoRS")
print("FWAS structurally predicted the VA cohort harm finding (HR 1.76) before unblinding.")
```
